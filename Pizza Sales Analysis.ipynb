```python
df['total_orders'] = df['order_id'].nunique()
```

```python
df.head()
```

```python

df['order_date'] = pd.to_datetime(df['order_date'])

# Extract day of the week
df['day_of_week'] = df['order_date'].dt.day_name()

# Calculate total distinct orders by day of the week
daily_orders = df.groupby('day_of_week')['order_id'].nunique()

# Order the days of the week
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_orders = daily_orders.reindex(days_order)
```

```python
plt.figure(figsize=(10, 6))
bar_plot = plt.bar(daily_orders.index, daily_orders.values, color='skyblue')

# Adding data labels
for bar, label in zip(bar_plot, daily_orders.values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label,
             ha='center', va='bottom', fontsize=10)

plt.title('Daily Trend of Total Distinct Orders')
plt.xlabel('Day of the Week')
plt.ylabel('Total Distinct Orders')
plt.show()
```

```python
# Extract month from the order_date
df['month'] = df['order_date'].dt.month_name()

# Calculate total distinct orders per month
monthly_orders = df.groupby('month')['order_id'].nunique()

# Plotting
plt.figure(figsize=(12, 6))
bar_plot = plt.bar(monthly_orders.index, monthly_orders.values, color='skyblue')

# Adding data labels
for bar, label in zip(bar_plot, monthly_orders.values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label,
             ha='center', va='bottom', fontsize=10)

plt.title('Monthly Trend of Total Distinct Orders')
plt.xlabel('Month')
plt.ylabel('Total Distinct Orders')
plt.show()
```

```python
df['order_time'] = pd.to_datetime(df['order_time'])
```

```python
# Extract the hour from the order_time
df['hour'] = df['order_time'].dt.hour

# Calculate total orders per hour
hourly_orders = df.groupby('hour')['order_id'].nunique()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(hourly_orders.index, hourly_orders.values, marker='o', linestyle='-', color='darkblue')

# Adding data labels
for i, value in enumerate(hourly_orders.values):
    plt.text(hourly_orders.index[i], value, str(value), ha='center', va='bottom', fontsize=10)

plt.title('Hourly Trend of Total Orders')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Orders')
plt.xticks(range(24))  # Set ticks for each hour
plt.grid(True)
plt.show()
```

```python
# Calculate total sales per pizza category
total_sales_per_category = df.groupby('pizza_category')['quantity'].sum()

# Display the result
print(total_sales_per_category)
```

```python
# Calculate total sales per pizza category
total_sales_per_category = df.groupby('pizza_category')['quantity'].sum()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(total_sales_per_category, labels=total_sales_per_category.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightblue'])
plt.title('Distribution of Sales Across Pizza Categories')
plt.show()
```

```python
# Calculate total sales per pizza size
total_sales_per_size = df.groupby('pizza_size')['quantity'].sum()

# Display the result
print(total_sales_per_size)
```

```python
# Calculate total sales per pizza size
total_sales_per_size = df.groupby('pizza_size')['quantity'].sum().reset_index()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(total_sales_per_size['quantity'], labels=total_sales_per_size['pizza_size'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Sales Across Pizza Sizes')
plt.show()
```

```python
# Calculate total orders according to the quantity of different pizza names
total_orders_per_pizza = df.groupby('pizza_name')['quantity'].sum().reset_index()

# Sort the DataFrame by total orders in descending order
total_orders_per_pizza = total_orders_per_pizza.sort_values(by='quantity', ascending=False)

# Select the top 5 best-selling pizzas
top5_pizzas = total_orders_per_pizza.head(5)

# Plotting
plt.figure(figsize=(8, 6))
bars = plt.bar(top5_pizzas['pizza_name'], top5_pizzas['quantity'], color='darkgreen')

# Adding data labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Best-Selling Pizzas Based on Total Orders')
plt.xlabel('Pizza Type')
plt.ylabel('Total Orders')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
